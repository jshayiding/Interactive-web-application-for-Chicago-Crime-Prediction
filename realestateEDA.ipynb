```python
import pandas as pd
import numpy as np
import seaborn as sns
import requests
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
api_key = None
with open('key.txt', 'r') as f:
    api_key = f.read().replace('\n','')
url_params = {'zws-id':api_key, 'state':'il', 'city':'chicago', 'childtype':'neighborhood'}
res = requests.get('https://www.zillow.com/webservice/GetRegionChildren.htm', params=url_params)
```

```python
import xml.etree.ElementTree as ET
tree = ET.fromstring(res.text)
```

```python
for child in tree:
    print(child.tag, child.attrib)
```

```python
neighborhoods = tree[2][2]
tags = []
for item in neighborhoods.findall('region'):
    for child in item:
        tags.append(child.tag)
tagset = set(tags)
```

```python
regions = []
for item in neighborhoods.findall('region'):
    region = {}
    childtags = []
    for child in item:
        region[child.tag] = child.text
        childtags.append(child.tag)
    if tagset.difference(set(childtags)) != set():
        for t in tagset.difference(set(childtags)):
            region[t] = None
    regions.append(region)
regions
```

```python
import csv
with open('zillow.csv', 'w', newline='') as csvfile:
    fields = list(tagset)
    writer = csv.DictWriter(csvfile, fieldnames=fields)
    writer.writeheader()
    writer.writerows(regions)
```
