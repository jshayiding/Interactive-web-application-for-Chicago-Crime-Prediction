```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import datetime
from scipy import stats
import seaborn as sns
```

```python
Real_estate_data = pd.read_csv('Real_estate.csv', sep=',', error_bad_lines=False)
# crimes_2012.drop(crimes_2012.columns[[0, 2, 5]], axis=1)
```

```python
Real_estate_data.apply(lambda x: sum(x.isnull()))
Real_estate_data.apply(lambda x: len(x.unique()))
```

```python
Real_estate_data.drop(['Jun-12','Feb-12','Mar-12','Feb-19', 'Apr-12'], inplace = True, axis = 1)
Real_estate_data = Real_estate_data.dropna(axis = 0, how = 'any')
```

```python
Real_estate_data['May-18'] = pd.Categorical(Real_estate_data['May-18'])
Real_estate_data['Nov-18'] = pd.Categorical(Real_estate_data['Nov-18'])
```

```python
## house price range for no of years

real_type_df = Real_estate_data['Region'].value_counts(ascending=True)

## Some formatting to make it look nicer
fig=plt.figure(figsize=(22, 20))
plt.title("Frequency of prices by region")
plt.xlabel("Frequency of prices")
plt.ylabel("Names of region")
ax = real_type_df.plot(kind='barh')
ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))


#plot of the relatives amounts of each type of crime
# plt.figure(figsize = (8,12))
# crimes_2012.groupby(crimes_2012['Primary Type']).size().sort_values(ascending = True).plot(kind = 'barh')
# plt.xlabel('Number of Crimes Committed')
# plt.show()
```

```python
Real_Data_date = Real_estate_data.pivot_table('Dec-18', aggfunc = np.size, columns = 'Region', 
                                         index = Real_estate_data.index.region, fill_value = 0)
Real_Data_date.index = pd.DatetimeIndex(Real_Data_date.index)

Plot = Real_Data_date.plot(figsize = (20,30), subplots = True, layout = (6,6),
                                  sharex = False, sharey = False)
plt.show()
```
